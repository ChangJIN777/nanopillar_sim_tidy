```python
# standard python imports
import numpy as np
from numpy import random
import matplotlib.pyplot as plt

import tidy3d as td
from tidy3d import web
from nanopillar_sim_base import *

nm = 1e-3
```

## setup batch sweep

```python
nanopillar_obj = nanopillar()
nanopillar_obj.set_runTime = 6e-12
nanopillar_obj.set_t_start = 5e-13
pillar_diameter_min = 200 # nm 
pillar_diameter_max = 600 # nm
diameter_step_size = 40
pillar_diameter_list = np.arange(pillar_diameter_min,pillar_diameter_max,diameter_step_size)
print(pillar_diameter_list)
sims = {f'pillar width={pillar_diameter:d}' : nanopillar_obj.sweep_pillar_width(pillar_diameter) for pillar_diameter in pillar_diameter_list}
batch = web.Batch(simulations=sims)
```

```python
batch_results = batch.run(path_dir='./data/nanopillar_sweep_1')
```

```python
i = 0 
for task_name, sim_data in batch_results.items():
    pillar_diameter = pillar_diameter_list[i]
    print(pillar_diameter)
    nanopillar_obj.plot_field_data(sim_data)
    nanopillar_obj.set_pillar_diameter(pillar_diameter)
    nanopillar_obj.data_analysis(sim_data)
    i = i+1
```
