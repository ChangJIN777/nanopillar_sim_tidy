```python
# standard python imports
import numpy as np
from numpy import random
import matplotlib.pyplot as plt

import tidy3d as td
from tidy3d import web
from nanopillar_sim_base import *

nm = 1e-3
```

```python
nanopillar_obj = nanopillar()
nanopillar_obj.set_runTime = 6e-12
nanopillar_obj.set_t_start = 5e-13
pillar_diameter = 200 # nm 
nanopillar_sim = nanopillar_obj.sweep_pillar_width(pillar_diameter)
```

```python
nanopillar_obj.extimate_cost(nanopillar_sim)
```

```python
nanopillar_data = nanopillar_obj.web_runSim(nanopillar_sim)
```

```python
nanopillar_obj.plot_field_data(nanopillar_data)
nanopillar_obj.data_analysis(nanopillar_data)
```

```python
top_monitor, bottom_monitor = nanopillar_obj.plot_flux_data(nanopillar_data)
data_summary = np.array([nanopillar_obj.freqs,
                                    top_monitor,
                                    bottom_monitor])
data_summary = data_summary.T
full_path_name = nanopillar_obj.data_path + 'nanopillar_diameter_%.0fnm.cvs'%(nanopillar_obj.W)
# adding the headline 
headline = "freq,top_monitor,bottom_monitor"
rows = ["{},{},{}".format(i, j, k) for i, j, k in data_summary]
text = "\n".join(rows)
text = headline + "\n" + text
with open(full_path_name,'w') as f:
    f.write(text)
```
